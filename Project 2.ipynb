```python
cols=['season','hr','holiday','weekday','weathersit','temp','hum','windspeed','casual','registered','cnt']
concat_data[cols]=concat_data[cols].apply(pd.to_numeric)
```

```python
q1=concat_data['temp'].quantile(.25)
q3=concat_data['temp'].quantile(.75)
```

```python
def find_outliers(df, cols):
    q1 = df[cols].quantile(.25)
    q3 = df[cols].quantile(.75)
    iqr = q3 - q1
    min_r = q1 - 1.5*iqr
    max_r = q3 + 1.5*iqr
    ouliers_indices = df.index[(df[cols] < min_r) | (df[cols] > max_r)]
    return ouliers_indices
```

```python
find_outliers(concat_data,'cnt')
```

```python
find_outliers(concat_data,'temp')
```

```python
find_outliers(concat_data,'hum')
```

```python
find_outliers(concat_data,'weathersit')
```
